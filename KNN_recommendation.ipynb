#Preparing the KNN

K Nearest Neighbor is a supervised machine learning algorithm but it can be used for unsupervised tasks as well. In this algorithm, no work is done on the training data hence training data is just stored in the memory.

When the testing data comes, its distance is calculated with all the training data points to find the k nearest neighbours which make this algorithm computationally very expensive.

We need a fast and optimized method to find the nearest neighbours. Using a 'brute structure' can speed up the search process.

In the context of a movie recommendation system using k-Nearest Neighbors (KNN) with the brute-force algorithm, you typically calculate the similarity between movies based on some features (e.g., genre, ratings, etc.) using a distance metric, such as cosine similarity.

```python
from sklearn.neighbors import NearestNeighbors
```

```python
#knn = NearestNeighbors(n_neighbors=6,algorithm='ball_tree').fit(movies)
```

```python
# Create a NearestNeighbors model using cosine similarity
knn_cosine = NearestNeighbors(n_neighbors=6, algorithm='brute', metric='cosine')
knn_cosine.fit(movies)
```

Brute Force kNN algorithm calculates the squared distances from each query feature vector to each reference feature vector in the training data set. Then, for each query feature vector it selects
 objects from the training set that are closest to that query feature vector.

```python
distances, indices = knn_cosine.kneighbors(movies)
```



*   distances: An array where each element distances[i] contains the distances between the i-th data point and its k nearest neighbors.
*   indices: An array where each element indices[i] contains the indices of the k nearest neighbors of the i-th data point.
This information can be useful, for example, in recommendation systems where you want to find items (movies in this case) that are similar to a given item based on certain features or characteristics.

```python
def improved_recommendations(movie_title, indices, distances, new_df):
    recommended_movies_list = []
    similar_movies_info = []

    # Check if the movie title exists in the dataset
    if movie_title not in new_df['name'].values:
        return f"Movie '{movie_title}' not found in the dataset."

    # Get the index of the input movie title
    movie_index = new_df[new_df['name'] == movie_title].index[0]

    # Get detailed information about the recommended movies
    for neighbor_id, distance in zip(indices[movie_index][1:], distances[movie_index][1:]):
        recommended_movie = new_df.loc[neighbor_id, ['name', 'genre','home', 'views', 'votes']]
        similar_movies_info.append({
            'name': recommended_movie['name'],
            'genre': recommended_movie['genre'],
            'home': recommended_movie['votes'],
            'views': recommended_movie['views'],
            'votes': recommended_movie['votes'],
            'similarity_score': 1 - distance
        })

    # Append the information to the list, including the original movie
    recommended_movies_list.append({
        'original_movie': movie_title,
        'similar_movies': similar_movies_info
    })

    # Create DataFrame from the accumulated data
    recommended_movies_df = pd.DataFrame(recommended_movies_list)

    return recommended_movies_df
```

```python
# Example usage
movie_title = "Пять ночей у Фредди (2023)"
improved_recommendations(movie_title, indices, distances, new_df)
```

```python
def improved_recommendations(movie_title, indices, distances, new_df):
    recommended_movies_list = []

    # Check if the movie title exists in the dataset
    if movie_title not in new_df['name'].values:
        return f"Movie '{movie_title}' not found in the dataset."

    # Get the index of the input movie title
    movie_index = new_df[new_df['name'] == movie_title].index[0]

    # Get detailed information about the recommended movies
    for neighbor_id, distance in zip(indices[movie_index][1:], distances[movie_index][1:]):
        recommended_movie = new_df.loc[neighbor_id, ['name', 'genre', 'views', 'home', 'votes']]
        #recommended_movie = new_df.loc[neighbor_id, ['name_eng', 'genre', 'views', 'home', 'votes']]
        recommended_movie['similarity_score'] = 1 - distance
        recommended_movies_list.append(recommended_movie)

    # Create DataFrame from the accumulated data
    recommended_movies_df = pd.DataFrame(recommended_movies_list)

    return recommended_movies_df

```

```python
# Example usage
movie_title = "Матрица (1999)"
improved_recommendations(movie_title, indices, distances, new_df)
```

#Testing the recommendation system


```python
# Sample 5000 movie titles from your DataFrame
movie_titles_list = df['name'].sample(n=5000, random_state=42).tolist()
```

```python
# Create empty lists to store the recommendations and related information
recommended_movies_list = []
views_list = []
votes_list = []
similarity_scores_list = []
mean_views_list = []
mean_votes_list = []
mean_similarity_scores_list = []

# Loop through each movie title
for movie_title in movie_titles_list:
    # Call the recommendation function for each movie title
    recommendations = improved_recommendations(movie_title, indices, distances, new_df)

    # Extract information from the recommendations
    recommended_movies_list.append(recommendations['name'].tolist())
    views_list.append(recommendations['views'].tolist())
    votes_list.append(recommendations['votes'].tolist())

    # Calculate mean of views and votes for the recommended movies
    mean_views = recommendations['views'].mean()
    mean_votes = recommendations['votes'].mean()
    mean_views_list.append(mean_views)
    mean_votes_list.append(mean_votes)

    # Extract similarity scores from the recommendations
    similarity_scores = recommendations['similarity_score'].tolist()
    similarity_scores_list.append(similarity_scores)

    # Calculate mean similarity score
    mean_similarity_score = sum(similarity_scores) / len(similarity_scores)
    mean_similarity_scores_list.append(mean_similarity_score)

# Create a new DataFrame with the results
recommendations_df = pd.DataFrame({
    'Original_Movie_Title': movie_titles_list,
    'Recommended_Movies': recommended_movies_list,
    'Recommended_Views': views_list,
    'Recommended_Votes': votes_list,
    'Mean_Views': mean_views_list,
    'Mean_Votes': mean_votes_list,
    'Similarity_Scores': similarity_scores_list,
    'Mean_Similarity_Score': mean_similarity_scores_list
})

```
