#Weighted_rating calculating

 Calculate the KP_metric weighted_rating() for each movie as a new feature score

wr.png

where,



*   v is the number of votes for the movie;
*   m is the minimum votes required to be listed in the chart;
*   R is the average rating of the movie
*   C is the mean vote across the whole report











KP_rating for all the movies is approx 6 on a scale of 10

```python
C= df['KP_rating'].mean()
C
```

m, the minimum votes required to be listed in the chart

```python
m= df['KP_votes'].quantile(0.9)
m
```

Here the movies for main Top list are filtering

```python
KP_movies = df.copy().loc[df['KP_votes'] >= m]
KP_movies.shape
```

```python
KP_movies
```

```python
def KP_weighted_rating(x, m=m, C=C):
    v = x['KP_votes']
    R = x['KP_rating']
    # Calculation is based on the IMDB formula
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
# Define a feature 'score' and calculate its value with `weighted_rating()`
KP_movies['score'] = KP_movies.apply(KP_weighted_rating, axis=1)
```

```python
# Sort movies by 'score' and 'year' columns in descending order
KP_movies = KP_movies.sort_values(by=['score', 'year'], ascending=[False, True])

# Select the top 20 movies
Top_movies_by_year = KP_movies[['name_eng', 'score', 'year', 'home_eng']].head(10)
```

```python
def build_chart_for_df(df, genre, percentile=0.85):
    # Filter the DataFrame by the specified genre
    df_genre = df[df['genre'].str.contains(genre, case=False, na=False)]

    # Extract vote counts and averages, and calculate C (mean of vote averages) and m (quantile of vote counts)
    vote_counts = df_genre[df_genre['votes'].notnull()]['votes'].astype('int')
    vote_averages = df_genre[df_genre['KP_rating'].notnull()]['KP_rating'].astype('float')
    C = vote_averages.mean()
    m = vote_counts.quantile(percentile)

    # Filter qualified movies based on conditions and select relevant columns
    qualified = df_genre[(df_genre['votes'] >= m) & (df_genre['votes'].notnull()) & (df_genre['KP_rating'].notnull())][['name', 'year', 'votes', 'KP_rating', 'genre', 'director', 'actors']]
    qualified['votes'] = qualified['votes'].astype('int')
    qualified['KP_rating'] = qualified['KP_rating'].astype('float')

    # Calculate the weighted rating (wr) using a simplified formula
    qualified['wr'] = qualified.apply(lambda x: (x['votes'] / (x['votes'] + m) * x['KP_rating']) + (m / (m + x['votes']) * C), axis=1)

    # Sort the movies by the weighted rating in descending order and select the top
    qualified = qualified.sort_values('wr', ascending=False).head(10)

    return qualified
```
